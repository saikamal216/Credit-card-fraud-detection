# Credit Card Fraud Detection

# Importing Libraries

```python
# Importing Libraries

# Data loading and Pre-processing
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Model
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
```

# Data Loading & Pre-processing

```python
# Data loading
data = pd.read_csv("/content/CC.csv")
```

```python
# Viewing top 5 rows of dataset
data.head()
```

```python
# Data Information
data.info()
```

```python
# Drop Unnecessary columns
data.drop('Unnamed: 0', axis=1, inplace=True)
```

# Data Visualization

```python
# Countplot for different labels
sns.countplot(x='Class', data=data)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

# Feature Enginerring

```python
# Separate majority and minority classes
df_0 = data[data.Class == 0]
df_1 = data[data.Class == 1]
```

```python
# Downsample majority class
df_0_downsampled = resample(df_0,
                                 replace=False,
                                 n_samples=len(df_1),
                                 random_state=123)
```

```python
# Combine minority class with downsampled majority class
df = pd.concat([df_0_downsampled, df_1])
```

```python
# Display new class counts
df.Class.value_counts()
```

```python
# Countplot for different labels after sampling
sns.countplot(x='Class', data=df)
plt.show()
```

```python
# Spliting Target column and feature column
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

# Models

Decision Tree CLassifier

```python
# Create a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
```

```python
# Train the model
dt_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_dt = dt_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", accuracy_dt)
```

```python
# Confusion matrix
cm_dt = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix:\n", cm_dt)
```

```python
# Classification report
cr_dt = classification_report(y_test, y_pred_dt)
print("Classification Report:\n", cr_dt)
```

Random Forest Classifier Model

```python
# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
rf_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Print a classification report
print(classification_report(y_test, y_pred))
```

Graident Boosting Classifier Model

```python
# Initialize the Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
gb_classifier.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_gb = gb_classifier.predict(X_test)
```

```python
# Evaluate the model
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Gradient Boosting Accuracy:", accuracy_gb)
```

```python
# Print a classification report
print(classification_report(y_test, y_pred_gb))
```

```python
models = ['Decision Tree', 'Random Forest', 'Gradient Boosting']
accuracies = [accuracy_dt, accuracy, accuracy_gb]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['blue', 'green', 'red'])
plt.title('Model Accuracies')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.show()
```
